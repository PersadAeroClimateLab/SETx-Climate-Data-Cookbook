# SETX-UIFL 1km Climate Atlas Overview

This dataset provides a daily evolution of meteorological conditions for the Southeast Texas region from **1950 through 2100**. It is designed to provide comprehensive coverage of potential climate outcomes for local stakeholders. Below the interactive atlas, you will find more background information on the data, how it was generated, and how to get more information.

**Dataset Composition:**
* **Temporal Coverage:** Daily data from 1950 to 2100.
* **Scenarios:** 4 different future climate scenarios (SSPs 1-2.6, 2-4.5, 3-7.0, and 5-8.5).
* **Model Diversity:** Realizations from 10 different global climate models.

**Key Variables:**
* **Temperature:** Daily Average, Maximum, and Minimum.
* **Humidity:** Daily Average Relative and Specific Humidity.
* **Wind & Energy:** Daily Average Near-Surface Wind Speed, Surface Solar Radiation, and Surface Longwave Radiation.

The data is made available on the Texas Advanced Computing Center's primary data-management and storage system [Corral](https://tacc.utexas.edu/systems/corral/) via a [CKAN interface and API](https://ckan.tacc.utexas.edu/). This notebook covers the following:

1. Understanding the available variables, scenarios, and models
2. Visualizing the spatial resolution and extent of the downscaled data
3. Dataset structure and preview of metadata

# Background Information on Climate Data

## Original Data Source: Global Climate Models (GCMs)

Our climate data begins with **Global Climate Models (GCMs)**. These are massive computer programs that simulate the behavior of the entire Earth's climate system—including the atmosphere, oceans, and land—based on fundamental physical equations. To do this, GCMs must divide the 3-dimensional climate system into cubes, often referred to as "grids", of a large enough size to be calculated efficiently on today's supercomputers.

* **A Global Effort:** Over 40 different GCMs have been developed by scientific institutions worldwide (such as **NASA, NOAA, and NSF**), each using slightly different approaches to represent complex climate processes.
* **Coordinated Simulations:** These models produce simulations of the historical period (to validate the models' performance at recreating observed climate conditions) and a standardized set of future scenarios, allowing scientists to compare results across different modeling groups.
* **A Targeted SETX Selection:** Each GCM performs better or worse at simulating specific things, depending on the choices made by the modeling center that built it. For example, certain models might be very good at simulating Pacific Ocean conditions, while others might be better at simulating climate over deserts. For our dataset, we selected 10 GCMs that performed best at representing he climate conditions that control flooding and air pollution over Southeast Texas. Details of this targeted SETX model selection process are provided in a forthcoming manuscript (Nduka et al. 2026).

For more information on global climate models, see:
* https://www.climate.gov/maps-data/climate-data-primer/predicting-climate/climate-models
* https://www.gfdl.noaa.gov/climate-modeling/

## Future Scenarios: Shared Socio-economic Pathways (SSPs)

Because we cannot know exactly how future society will change, researchers use **Shared Socio-economic Pathways (SSPs)** to explore different "what-if" scenarios. These scenarios represent different trajectories of greenhouse gas and other air pollutant emissions and additional societal choices, like land use, that affect the climate.

* **The Range of Severities:**
    - **SSP1-2.6: Low climate change severity**
      - **The Story:** A world moving toward sustainability, with a focus on human well-being, clean energy, and international cooperation.
      - **The Result:** A low-emissions future that aligns with the Paris Agreement goal of keeping global warming below 2°C.
    - **SSP2-4.5: Intermediate climate change severity**
      - **The Story:** A future where social, economic, and technological trends follow historical patterns with slow progress toward sustainability.
      - **The Result:** An intermediate emissions scenario where global temperatures continue to rise but at a more moderate pace than the "worst-case" projections.
    - **SSP3-7.0: High climate change severity**
      - **The Story:** A world characterized by resurgent nationalism, regional conflicts, and a focus on domestic security rather than global climate goals.
      - **The Result:** A high-emissions scenario with significant challenges to both preparing for (adapting to) and reducing (mitigating) climate change.
    - **SSP5-8.5: Highest climate change severity**
      - **The Story:** Rapid global economic growth driven by an intensive use of fossil fuels and energy-heavy lifestyles.
      - **The Result:** The highest-emissions scenario used in climate modeling, often used to understand "worst-case" climate impacts and extreme weather possibilities.

* **Range of Potential Futures:** There is no "most likely" scenario; rather, the SSPs capture the full range of potential futures based on global policy and energy choices.

For more information on the Shared Socio-economic Pathways (SSPs) see: 
* https://climatedata.ca/resource/understanding-shared-socio-economic-pathways-ssps/
* https://earth.gov/sealevel/faq/124/what-are-shared-socioeconomic-pathways-or-ssps/

## From Global to Local: The 1km Downscaling Process

While global climate models (GCMs) are powerful, the grids that they use to calculate climate conditions worldwide are often too coarse for city-scale planning. A typical GCM grid cell is roughly 100km, which is larger than the entire Beaumont-Port Arthur area. To make this data useful for Southeast Texas, we perform **spatial downscaling**.

* **The Input:** We start with the **NASA Earth Exchange Global Daily Downscaled Projections (NEX-GDDP-CMIP6)** dataset, which provides global coverage at a 0.25° (~25km) resolution. This dataset it generated by the U.S. National Aeronautics and Space Administration, and more information on it can be found here: https://www.nccs.nasa.gov/services/data-collections/land-based-products/nex-gddp-cmip6
* **The Result:** Using advanced statistical and physical techniques, we "downscale" this information to a **1km spatial resolution**. This allows us to see how climate impacts vary across the Southeast Texas region. We do this by relating the coarse resolution climate data with 1km spatial resolution observations to establish statistical relationships between coarse-resolution and 1km-resolution climate behaviors. We then apply these statistical relationships across the available climate projection data to translate it from ~25km to 1km. The full documentation of this process is provided in a forthcoming manuscript (Nduka et al. 2026).

# How to Run the Atlas

### To start using the Atlas:
1. Find the `Run` menu at the top of this screen, and select `Run All Cells`.

2. Once the data download is complete, scroll down the page and you will see the interactive atlas.

>*Note: the atlas will take a few minutes to fetch the data that will be used. The atlas will display the progress and the estimated time to completion of this step.*

### Navigating the Atlas:

1. Dropdown menus at the top of the interactive atlas allow you to select the meteorological variable, future climate scenario, future time period, and range of climate models (all or a single climate model) for which data will be shown on the map figure. You can also select whether you want data for all counties or only for a specific county to be shown on the graph figure that visualizes how conditions are changing through time.
2. On the map figure, you can hover over individual counties  to see the average value of conditions depicted on the map for that county.
3. On the graph figure, you can select and deselect individual models by clicking on their name in the legend to the right of the figure. You can also hover over the graph to see the values for each year, each selected model and the median value among all models.

>*Note: Users considering next-step applications of the dataset should keep in mind that, although the atlas shows data aggregated to the county level, the underlying data provides 1 km spatial resolution.*
