## Dataset Structure

All data is stored in NetCDF format with some selections additionally made available as Zarr stores. NetCDF files are chunked by year and average about 1 GB in size. The data hirearchy is visaulized below for an individual NetCDF file:

```
hurs/                                            # variable
└── hist/                                        # scenario
    └── CNRM-ESM2-1/                             # model
        └── CNRM-ESM2-1_hist_hurs_1km_1950.nc    # file.nc
```

This file can be downloaded via CKAN [here](https://ptdatax.tacc.utexas.edu/workbench/data/tapis/projects/ptdatax.project.PTDATAX-207/SETx_Climate_Datasets/hurs/hist/CNRM-ESM2-1/CNRM-ESM2-1_hist_hurs_1km_1950.nc/) (we can't automate or script the download since it requires credentials, you will have to download manually if you wish to inspect it further). Once downloaded, we can utilize `xarray` to open the dataset and view the metadata:

Two values are stored in each NetCDF file: the climate variable ("hurs") and the associated standard error ("hurs_SE"). Attributes describing the units, coordinates, and references can be inspected by clicking through the dropdown menus.

### Available Variables

There are 9 variables availabe in the dataset:

- tasmax – Near-surface maximum air temperature
- tasmin – Near-surface minimum air temperature
- tas – Near-surface mean air temperature
- pr – Daily precipitation total
- hurs – Near-surface relative humidity
- huss – Near-surface specific humidity
- rsds – Surface downwelling shortwave radiation
- rlds – Surface downwelling longwave radiation
- sfcWind – Near-surface wind speed

### Available Models

For each variable, there are 10 models:

- BCC-CSM2-MR
- CESM2
- CMCC-ESM2
- CNRM-ESM2-1
- EC-Earth3
- FGOALS-g3
- GFDL-CM4
- MPI-ESM1-2-HR
- MRI-ESM2-0
- NorESM2-MM


### Available Scenarios

For each variable and each model, there are 5 scenarios:

- Historical (1950 to 2014)
- SSP1-2.6 (2015 to 2100)
- SSP2-4.5 (2015 to 2100)
- SSP3-7.0 (2015 to 2100)
- SSP5-8.5 (2015 to 2100)

The future scenarios (2015 to 2100) follow the [Shared Socioeconomic Pathways (SSP)](https://ssp.apps.ece.iiasa.ac.at/) framework developed for the IPCC 6th Assessment Report.

## Exploring Annual Means

The code below creates an interactive plot that explores annual means taken across every combination of variable, scenario, and model available.

:::{note}
The full dataset is over 30 TB in size. This is too large for a single binder-based notebook to handle, so we have post-processed the data down to a collection of statistical metrics to make it easier to visualize. This was done efficiently using `xarray` support for `dask`, a parallelization library that allows us to leverage multiple cores on a computer to perform the heavy computations. Thus, the data referenced in the code below points to datasets *derived datasets* rather than the original full dataset. Future notebooks will explore how to effectively analyze such a large dataset, but for now we will rely on the numbers crunched by our favorite high-performance computer `thunderhead-4`. 
:::

````python
historical_ds = pd.read_csv("../post_proc_data/setx_post_proc_historical_yr_timeseries.csv", index_col=0)
future_ds = pd.read_csv("../post_proc_data/setx_post_proc_future_yr_timeseries.csv", index_col=0)

future_ds
````

:::{seealso}
You can view the Python code we used to compute these post-processed datasets in `post_proc_data/compute_yr_timeseries.py`
:::

````python
models = ['BCC-CSM2-MR', 'CESM2', 'CMCC-ESM2', 'CNRM-ESM2-1', 'EC-Earth3', 'FGOALS-g3', 'GFDL-CM4', 'MPI-ESM1-2-HR', 'MRI-ESM2-0', 'NorESM2-MM', 'UKESM1-0-LL']
scenarios = ['ssp126', 'ssp245', 'ssp370', 'ssp585', 'hist_avg']
variables = ['hurs', 'huss', 'rlds', 'rsds', 'sfcWind', 'tas', 'tasmax', 'tasmin']

labels = {
    'hurs': "Relative Humidity (%)",
    'huss': "Specific Humidity (kg/kg)",
    'rlds': "Downwelling Longwave Radiation (W/m-2)",
    'rsds': "Downwelling Shortwave Radiation (W/m-2)",
    'sfcWind': "Near-Surface Wind Speed (m/s)",
    'tas': "Temperature (°F)",
    'tasmax': "Temperature (°F)",
    'tasmin': "Temperature (°F)"
}

future_ds = pd.read_csv("../post_proc_data/setx_post_proc_future_yr_timeseries.csv", index_col=0)
future_years = [int(date.split("-")[0]) for date in future_ds.index]
model = models[0]
variable = "tas"

new_ds = {}
for entry in historical_ds:
    entry_model, tag, entry_var = entry.split("_")
    new_ds[f"{entry_model}_hist_avg_{entry_var}"] = historical_ds[entry].mean()*np.ones(len(future_years))
future_ds = pd.concat([pd.DataFrame(new_ds, index=future_ds.index), future_ds], axis=1)

fig = go.FigureWidget()
fig.update_layout(
    template="plotly_white",
    xaxis_title="Year",
    yaxis_title=labels[variable],
    legend_title="Scenario",
    width=1200, height=800
)

fig.update_layout(title=f"{model.upper()} Projected Annual Mean {variable.upper()} by Scenario")
fig.update_layout(
    font=dict(
        family="Inter",
        size=20,
    )
)

for index, ssp in enumerate(scenarios):
    data_str = f"{model}_{ssp}_{variable}"
    color = None
    if ssp == "hist_avg":
        color = "black"
    if data_str in future_ds:
        data = future_ds[data_str].values
        if "tas" in variable:
            data = (data - 273.15) * (9.0 / 5.0) + 32.0
        fig.add_scatter(x=future_years, y=data, mode="lines", name=ssp, marker=dict(
            color=color
        ))
        fig.data[index].visible = True
    else:
        fig.add_scatter(x=future_years, y=np.zeros(len(future_years)), mode="lines", name=ssp)
        fig.data[index].visible = False

def on_variable_change(change):
    variable = change["new"]
    with fig.batch_update():
        for index, ssp in enumerate(scenarios):
            data_str = f"{model}_{ssp}_{variable}"
            if data_str in future_ds:
                data = future_ds[data_str].values
                if "tas" in variable:
                    data = (data - 273.15) * (9.0 / 5.0) + 32.0
                elif "hurs" == variable:
                    data = 100*data
                fig.data[index].y = data
                fig.data[index].visible = True
            else:
                fig.data[index].visible = False
        fig.layout.title.text = f"{model.upper()} Projected Annual Mean {variable.upper()} by Scenario"
        fig.update_layout(yaxis_title=labels[variable])

variable_toggle = widgets.ToggleButtons(
    value=variable,
    options=variables,
    button_style="info",
)
variable_toggle.observe(on_variable_change, names="value")

def on_model_change(change):
    model = change["new"]
    with fig.batch_update():
        for index, ssp in enumerate(scenarios):
            data_str = f"{model}_{ssp}_{variable}"
            if data_str in future_ds:
                data = future_ds[data_str].values
                if "tas" in variable:
                    data = (data - 273.15) * (9.0 / 5.0) + 32.0
                elif "hurs" == variable:
                    data = 100*data
                fig.data[index].y = data
                fig.data[index].visible = True
            else:
                fig.data[index].visible = False
        fig.layout.title.text = f"{model.upper()} Projected Annual Mean {variable.upper()} by Scenario"

model_toggle = widgets.ToggleButtons(
    value=model,
    options=models,
    button_style="info",
)
model_toggle.observe(on_model_change, names="value")

display(variable_toggle, model_toggle, fig)
````
