## Welcome to the Southeast Texas Urban Integrated Field Laboratory (SETX-UIFL) 1km Climate Atlas Cookbook

This interactive guide serves as a gateway to the **SETX-UIFL 1km Climate Atlas**, a high-resolution projection dataset specifically designed for the unique coastal, urban, rural, industrial, and natural landscapes of Southeast Texas. By downscaling global climate models to a granular **1 km spatial resolution**, we provide the community, stakeholders, and researchers with the tools needed to visualize and prepare for future environmental shifts.

> **Note:** This cookbook is part of the [Southeast Texas UIFL](https://setx-uifl.org/), supported by the U.S. Department of Energy.

# SETX-UIFL 1km Climate Atlas Overview

This dataset provides a daily evolution of meteorological conditions for the Southeast Texas region from **1950 through 2100**. It is designed to provide comprehensive coverage of potential climate outcomes for local stakeholders.

**Dataset Composition:**
* **Temporal Coverage:** Daily data from 1950 to 2100.
* **Scenarios:** 4 different future climate scenarios (SSPs 1-2.6, 2-4.5, 3-7.0, and 5-8.5).
* **Model Diversity:** Realizations from 10 different global climate models.

**Key Variables:**
* **Temperature:** Daily Average, Maximum, and Minimum.
* **Humidity:** Daily Average Relative and Specific Humidity.
* **Wind & Energy:** Daily Average Near-Surface Wind Spee, Surface Solar Radiation, and Surface Longwave Radiation.

The data is made available on the Texas Advanced Computing Center's primary data-management and storage system [Corral](https://tacc.utexas.edu/systems/corral/) via a [CKAN interface and API](https://ckan.tacc.utexas.edu/). This notebook covers the following:

1. Understanding the available variables, scenarios, and models
2. Visualizing the spatial resolution and extent of the downscaled data
3. Dataset structure and preview of metadata

# Background Information on Climate Data

## Original Data Source: Global Climate Models (GCMs)

Our climate data begins with **Global Climate Models (GCMs)**. These are massive computer programs that simulate the behavior of the entire Earth's climate system—including the atmosphere, oceans, and land—based on fundamental physical equations. To do this, GCMs must divide the 3-dimensional climate system into cubes, often referred to as "grids", of a large enough size to be calculated efficiently on today's supercomputers.

* **A Global Effort:** Over 40 different GCMs have been developed by scientific institutions worldwide (such as **NASA, NOAA, and NSF**), each using slightly different approaches to represent complex climate processes.
* **Coordinated Simulations:** These models produce simulations of the historical period (to validate the models' performance at recreating observed climate conditions) and a standardized set of future scenarios, allowing scientists to compare results across different modeling groups.
* **A Targeted SETX Selection:** Each GCM performs better or worse at simulating specific things, depending on the choices made by the modeling center that built it. For example, certain models might be very good at simulating Pacific Ocean conditions, while others might be better at simulating climate over deserts. For our dataset, we selected 10 GCMs that performed best at representing he climate conditions that control flooding and air pollution over Southeast Texas. Details of this targeted SETX model selection process are provided in a forthcoming manuscript (Nduka et al. 2026).

For more information on global climate models, see:
* https://www.climate.gov/maps-data/climate-data-primer/predicting-climate/climate-models
* https://www.gfdl.noaa.gov/climate-modeling/

## Future Scenarios: Shared Socio-economic Pathways (SSPs)

Because we cannot know exactly how future society will change, researchers use **Shared Socio-economic Pathways (SSPs)** to explore different "what-if" scenarios. These scenarios represent different trajectories of greenhouse gas and other air pollutant emissions and additional societal choices, like land use, that affect climate.

* **The Range of Severities:**
    - **SSP1-2.6: Low climate change severity**
      - **The Story:** A world moving toward sustainability, with a focus on human well-being, clean energy, and international cooperation.
      - **The Result:** A low-emissions future that aligns with the Paris Agreement goal of keeping global warming below 2°C.
    - **SSP2-4.5: Intermediate climate change severity**
      - **The Story:** A future where social, economic, and technological trends follow historical patterns with slow progress toward sustainability.
      - **The Result:** An intermediate emissions scenario where global temperatures continue to rise but at a more moderate pace than the "worst-case" projections.
    - **SSP3-7.0: High climate change severity**
      - **The Story:** A world characterized by resurgent nationalism, regional conflicts, and a focus on domestic security rather than global climate goals.
      - **The Result:** A high-emissions scenario with significant challenges to both preparing for (adapting to) and reducing (mitigating) climate change.
    - **SSP5-8.5: Highest climate change severtiy**
      - **The Story:** Rapid global economic growth driven by an intensive use of fossil fuels and energy-heavy lifestyles.
      - **The Result:** The highest-emissions scenario used in climate modeling, often used to understand "worst-case" climate impacts and extreme weather possibilities.

* **Range of Potential Futures:** There is no "most likely" scenario; rather, the SSPs capture the full range of potential futures based on global policy and energy choices.

For more information on the Shared Socio-economic Pathways (SSPs) see: 
* https://climatedata.ca/resource/understanding-shared-socio-economic-pathways-ssps/
* https://earth.gov/sealevel/faq/124/what-are-shared-socioeconomic-pathways-or-ssps/

## From Global to Local: The 1km Downscaling Process

While global climate models (GCMs) are powerful, their grids that they use the calculate climate conditions worldwide are often too coarse for city-scale planning. A typical GCM grid cell is roughly 100km, which is larger than the entire Beaumont-Port Arthur area. To make this data useful for Southeast Texas, we perform **spatial downscaling**.

* **The Input:** We start with the **NASA Earth Exchange Global Daily Downscaled Projections (NEX-GDDP-CMIP6)** dataset, which provides global coverage at a 0.25° (~25km) resolution. This dataset it generated by the U.S. National Aeronautics and Space Administration, and more information on it can be found here: https://www.nccs.nasa.gov/services/data-collections/land-based-products/nex-gddp-cmip6
* **The Result:** Using advanced statistical and physical techniques, we "downscale" this information to a **1km spatial resolution**. This allows us to see how climate impacts vary across the Southeast Texas region. We do this by relating the coarse resolution climate data with 1km spatial resolution observations to establish statistical relationships between coarse-resolution and 1km-resolution climate behaviors. We then apply these statistical relationships across the available climate projection data to translate it from ~25km to 1km. The full documentation of this process are provided in a forthcoming manuscript (Nduka et al. 2026).

````python
import numpy as np
import pandas as pd
import plotly.graph_objects as go
import ipywidgets as widgets
from IPython.display import display

models = ['BCC-CSM2-MR', 'CESM2', 'CMCC-ESM2', 'CNRM-ESM2-1', 'EC-Earth3', 'FGOALS-g3', 'GFDL-CM4', 'MPI-ESM1-2-HR', 'MRI-ESM2-0', 'NorESM2-MM', 'UKESM1-0-LL']
scenarios = ['ssp126', 'ssp245', 'ssp370', 'ssp585', 'hist_avg']
variables = ['hurs', 'huss', 'rlds', 'rsds', 'sfcWind', 'tas', 'tasmax', 'tasmin']
labels = {
    'hurs': "Relative Humidity (%)",
    'huss': "Specific Humidity (kg/kg)",
    'rlds': "Downwelling Longwave Radiation (W/m-2)",
    'rsds': "Downwelling Shortwave Radiation (W/m-2)",
    'sfcWind': "Near-Surface Wind Speed (m/s)",
    'tas': "Temperature (°F)",
    'tasmax': "Temperature (°F)",
    'tasmin': "Temperature (°F)"
}

# ─── IPCC AR6 scenario colors (from pyam / IPCC WGI Visual Style Guide) ───
scenario_colors = {
    'ssp126': '#003466',   # dark blue
    'ssp245': '#f69320',   # orange
    'ssp370': '#df0000',   # red
    'ssp585': '#980002',   # dark crimson
    'hist_avg': '#000000',  # black
}

# Friendly display labels for toggle buttons
variable_options = [
    ("Relative Humidity", "hurs"),
    ("Specific Humidity", "huss"),
    ("Downwelling Longwave Radiation", "rlds"),
    ("Downwelling Shortwave Radiation", "rsds"),
    ("Wind Speed", "sfcWind"),
    ("Daily Average Temperature", "tas"),
    ("Daily High Temperature", "tasmax"),
    ("Daily Low Temperature", "tasmin"),
]

scenario_options = [
    ("Least Severe (SSP1-2.6)", "ssp126"),
    ("Intermediate (SSP2-4.5)", "ssp245"),
    ("High Severity (SSP3-7.0)", "ssp370"),
    ("Most Severe (SSP5-8.5)", "ssp585"),
    ("Historical Average", "hist_avg"),
]

def get_variable_friendly(variable):
    for label, var_technical in variable_options:
        if variable == var_technical:
            return label
    return variable

def get_ssp_friendly(ssp):
    for label, ssp_technical in scenario_options:
        if ssp == ssp_technical:
            return label
    return ssp

# Load and prepare data
historical_ds = pd.read_csv("../post_proc_data/setx_post_proc_historical_yr_timeseries.csv", index_col=0)
future_ds = pd.read_csv("../post_proc_data/setx_post_proc_future_yr_timeseries.csv", index_col=0)
future_years = [int(date.split("-")[0]) for date in future_ds.index]

new_ds = {}
for entry in historical_ds:
    entry_model, tag, entry_var = entry.split("_")
    new_ds[f"{entry_model}_hist_avg_{entry_var}"] = historical_ds[entry].mean() * np.ones(len(future_years))
future_ds = pd.concat([pd.DataFrame(new_ds, index=future_ds.index), future_ds], axis=1)

# Helper to convert units
def convert_data(data, variable):
    if "tas" in variable:
        return (data - 273.15) * (9.0 / 5.0) + 32.0
    elif variable == "hurs":
        return 100 * data
    return data

# Initial selections
model = models[0]
variable = "tas"

# Build figure with one trace per scenario
fig = go.FigureWidget()
fig.update_layout(
    template="plotly_white",
    xaxis_title="Year",
    yaxis_title=labels[variable],
    legend_title="Scenario",
    width=1200, height=800,
    title=f"{model.upper()} Projected Annual Mean {get_variable_friendly(variable)} by Scenario",
    font=dict(family="Inter", size=20),
)

for ssp in scenarios:
    fig.add_scatter(
        x=future_years, y=[], mode="lines", name=get_ssp_friendly(ssp),
        line=dict(color=scenario_colors[ssp]), visible=False
    )

# Shared update function
def update_line_plot(*args):
    variable = variable_toggle.value
    model = model_toggle.value
    with fig.batch_update():
        for index, ssp in enumerate(scenarios):
            data_str = f"{model}_{ssp}_{variable}"
            if data_str in future_ds:
                fig.data[index].y = convert_data(future_ds[data_str].values, variable)
                fig.data[index].visible = True
            else:
                fig.data[index].visible = False
        fig.layout.title.text = f"{model.upper()} Projected Annual Mean {get_variable_friendly(variable)} by Scenario"
        fig.layout.yaxis.title.text = labels[variable]

# Line plot toggles
variable_toggle = widgets.ToggleButtons(
    value="tas", options=variable_options, description="Variable:", button_style="info"
)
model_toggle = widgets.ToggleButtons(
    value=models[0], options=models, description="Model:", button_style="info"
)

variable_toggle.observe(update_line_plot, names="value")
model_toggle.observe(update_line_plot, names="value")

# Initialize traces with data
update_line_plot()

display(variable_toggle, model_toggle, fig)
````

````python
import numpy as np
import xarray as xr
import geopandas as gpd
import plotly.graph_objects as go
import ipywidgets as widgets
from IPython.display import display
import warnings
warnings.filterwarnings("ignore", category=UserWarning)


models = ['BCC-CSM2-MR', 'CESM2', 'CMCC-ESM2', 'CNRM-ESM2-1', 'EC-Earth3', 'FGOALS-g3', 'GFDL-CM4', 'MPI-ESM1-2-HR', 'MRI-ESM2-0', 'NorESM2-MM', 'UKESM1-0-LL']
scenarios = ['ssp126', 'ssp245', 'ssp370', 'ssp585']
variables = ['hurs', 'huss', 'rlds', 'rsds', 'sfcWind', 'tas', 'tasmax', 'tasmin']
labels = {
    'hurs': "Relative Humidity (%)",
    'huss': "Specific Humidity (kg/kg)",
    'rlds': "Downwelling Longwave Radiation (W/m-2)",
    'rsds': "Downwelling Shortwave Radiation (W/m-2)",
    'sfcWind': "Near-Surface Wind Speed (m/s)",
    'tas': "Temperature (°F)",
    'tasmax': "Temperature (°F)",
    'tasmin': "Temperature (°F)"
}

# Initial selections
model = 'BCC-CSM2-MR'
variable = "tas"
scenario = "ssp370"

# Load shapefile once
borders = gpd.read_file("../post_proc_data/texas_shapefile/State_Boundary.shp").to_crs(crs=4326)

variable_options = [
    ("Relative Humidity", "hurs"),
    ("Specific Humidity", "huss"),
    ("Downwelling Longwave Radiation", "rlds"),
    ("Downwelling Shortwave Radiation", "rsds"),
    ("Wind Speed", "sfcWind"),
    ("Daily Average Temperature", "tas"),
    ("Daily High Temperature", "tasmax"),
    ("Daily Low Temperature", "tasmin"),
]

scenario_options = [
    ("Least Severe (SSP1-2.6)", "ssp126"),
    ("Intermediate (SSP2-4.5)", "ssp245"),
    ("High Severity (SSP3-7.0)", "ssp370"),
    ("Most Severe (SSP5-8.5)", "ssp585")
]

# ─── IPCC-style colormaps per variable (for spatial maps) ───
variable_colorscales = {
    'tas':     'RdBu_r',     # temperature: warm diverging
    'tasmax':  'RdBu_r',
    'tasmin':  'RdBu_r',
    'hurs':    'Teal',        # relative humidity: blue-green sequential
    'huss':    'Teal',        # specific humidity: blue-green sequential
    'rlds':    'Inferno',     # longwave radiation: dark-to-bright sequential
    'rsds':    'Plasma',      # shortwave radiation: cool-to-warm sequential
    'sfcWind': 'Viridis',     # wind speed: perceptually uniform sequential
}

def get_variable_friendly(variable):
    for label, var_technical in variable_options:
        if variable == var_technical:
            return label
    return variable

def get_ssp_friendly(ssp):
    for label, ssp_technical in scenario_options:
        if ssp == ssp_technical:
            return label
    return ssp

# Helper to load data
def load_map_data(variable, model, scenario):
    da = xr.open_dataset(
        f"../post_proc_data/SETx-UIFL-Climate-Downscaled-Post_realization-maps_{variable}.nc"
    )[variable].sel(realization=f"{model}-{scenario}")
    if "tas" in variable:
        da = (da - 273.15) * (9.0 / 5.0) + 32.0
    elif "hurs" == variable:
        da = 100*da
    return da

# Initial data
da = load_map_data(variable, model, scenario)

# Build figure
map_fig = go.FigureWidget(data=go.Heatmap(
    z=da.values,
    x=da.coords["longitude"].values,
    y=da.coords["latitude"].values,
    colorscale=variable_colorscales[variable],
    colorbar=dict(title=dict(text=labels[variable], side="right"))
))

map_fig.update_layout(
    template="plotly_white",
    xaxis_title="Longitude",
    yaxis_title="Latitude",
    width=1000,
    height=900,
    title=da.attrs.get("long_name", da.name),
    font=dict(family="Inter", size=20),
)

lon = da.coords["longitude"].values
lat = da.coords["latitude"].values

map_fig.update_layout(
    xaxis=dict(range=[lon.min(), lon.max()]),
    yaxis=dict(range=[lat.min(), lat.max()]),
)

# Add border traces
for _, row in borders.iterrows():
    geom = row.geometry
    polys = geom.geoms if geom.geom_type == "MultiPolygon" else [geom]
    for poly in polys:
        x, y = poly.exterior.coords.xy
        map_fig.add_trace(go.Scatter(
            x=np.array(x), y=np.array(y),
            mode="lines",
            line=dict(color="black", width=1),
            showlegend=False,
            hoverinfo="skip",
        ))

# Shared update function
def update_map(*args):
    variable = map_variable_toggle.value
    model = map_model_toggle.value
    scenario = map_scenario_toggle.value
    try:
        da = load_map_data(variable, model, scenario)
        lon = da.coords["longitude"].values
        lat = da.coords["latitude"].values

        with map_fig.batch_update():
            map_fig.data[0].z = da.values
            map_fig.data[0].x = da.coords["longitude"].values
            map_fig.data[0].y = da.coords["latitude"].values
            map_fig.data[0].colorscale = variable_colorscales[variable]
            map_fig.data[0].colorbar.title = labels[variable]
            map_fig.data[0].colorbar.title.side = "right"
            map_fig.layout.xaxis.range = [lon.min(), lon.max()]
            map_fig.layout.yaxis.range = [lat.min(), lat.max()]
            map_fig.layout.title.text = f"{model.upper()} {get_ssp_friendly(scenario)} — {get_variable_friendly(variable)}"
    except Exception as e:
        map_fig.layout.title.text = f"Data not available for {model}-{scenario}-{variable}: {e}"

# Toggle buttons
map_variable_toggle = widgets.ToggleButtons(
    value="tas", options=variable_options, description="Variable:", button_style="info"
)
map_model_toggle = widgets.ToggleButtons(
    value=models[0], options=models, description="Model:", button_style="info"
)
map_scenario_toggle = widgets.ToggleButtons(
    value="ssp126", options=scenario_options, description="Scenario:", button_style="info"
)

map_variable_toggle.observe(update_map, names="value")
map_model_toggle.observe(update_map, names="value")
map_scenario_toggle.observe(update_map, names="value")

update_map()

display(map_variable_toggle, map_model_toggle, map_scenario_toggle, map_fig)
````

## Summary of SETx Downscaled Daily 1km Climate Dataset

### Available Variables

There are 9 variables availabe in the dataset:

- tasmax – Near-surface maximum air temperature
- tasmin – Near-surface minimum air temperature
- tas – Near-surface mean air temperature
- pr – Daily precipitation total
- hurs – Near-surface relative humidity
- huss – Near-surface specific humidity
- rsds – Surface downwelling shortwave radiation
- rlds – Surface downwelling longwave radiation
- sfcWind – Near-surface wind speed

### Available Models

For each variable, there are 10 models:

- BCC-CSM2-MR
- CESM2
- CMCC-ESM2
- CNRM-ESM2-1
- EC-Earth3
- FGOALS-g3
- GFDL-CM4
- MPI-ESM1-2-HR
- MRI-ESM2-0
- NorESM2-MM


### Available Scenarios

For each variable and each model, there are 5 scenarios:

- Historical (1950 to 2014)
- SSP1-2.6 (2015 to 2100)
- SSP2-4.5 (2015 to 2100)
- SSP3-7.0 (2015 to 2100)
- SSP5-8.5 (2015 to 2100)

The future scenarios (2015 to 2100) follow the [Shared Socioeconomic Pathways (SSP)](https://ssp.apps.ece.iiasa.ac.at/) framework developed for the IPCC 6th Assessment Report.
